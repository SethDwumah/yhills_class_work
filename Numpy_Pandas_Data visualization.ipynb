```python
class Plot():
    
    import seaborn as sns
    import matplotlib.pyplot as plt
    plt.rcParams['figure.figsize'] = (6,4)
    
    def scatterPlot(self,df1,df2):
        sns.scatterplot(df1,df2) 
        plt.show()
        
    def histPlot(self,df1):
        sns.histplot(df1) 
        plt.show()
        
    def boxPlot(self,df1,df2):
        sns.boxplot(df1,df2) 
        plt.show()
        
    def countPlot(self,df1):
        sns.countplot(df1) 
        plt.show()
    
        
```

```python
plot =Plot()
```

```python
plot.scatterPlot(df['mpg'],df['displacement'])
```
